```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_video(video_path, resize=(224,224)):
    cap = cv2.VideoCapture(video_path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv2.resize(frame, resize)
        frames.append(frame)

    cap.release()
    return frames
```

```python
def motion_score(frames):
    diffs = []

    for i in range(len(frames)-1):
        f1 = cv2.cvtColor(frames[i], cv2.COLOR_BGR2GRAY)
        f2 = cv2.cvtColor(frames[i+1], cv2.COLOR_BGR2GRAY)

        diff = cv2.absdiff(f1, f2)
        diffs.append(np.mean(diff))

    return np.mean(diffs)
```

```python
def stage0_motion_trigger(video_path, window_size=32, threshold=6):
    cap = cv2.VideoCapture(video_path)
    buffer = []
    active_segments = []
    timestamps = []

    frame_id = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv2.resize(frame, (224,224))
        buffer.append(frame)

        if len(buffer) == window_size:
            score = motion_score(buffer)

            if score > threshold:
                active_segments.append(buffer.copy())
                timestamps.append(frame_id)

            buffer.pop(0)

        frame_id += 1

    cap.release()
    return active_segments, timestamps
```

```python
video_path = "test_video.mp4"
segments, times = stage0_motion_trigger(video_path)

print("Active segments found:", len(segments))
print("Frame positions:", times)
```

```python
def show_segment(segment):
    for f in segment[:10]:
        plt.imshow(cv2.cvtColor(f, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()

if len(segments) > 0:
    show_segment(segments[0])
```

```python
cap = cv2.VideoCapture(0)  
buffer = []

while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, (224,224))
    buffer.append(frame)

    if len(buffer) > 32:
        score = motion_score(buffer[-32:])

        if score > 6:
            print("Motion detected:", score)

        buffer.pop(0)

    cv2.imshow("Stage-0", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
